```python
# "돈이 있으면 물건을 구매할 수 있고, 없으면 구매할 수 없다."를 컴퓨터는 어떻게 이해할까요?

money = True

if money:
    print("물건을 구매할 수 있다.")
else:
    print("물건을 구매할 수 없다.")
```

```python
# 5000원 이상이 있으면 커피를 마실 수 있고, 없으면 커피를 마시기에 돈이 부족하다.

money = 6000

if money >= 5000:
    print("커피를 마실 수 있다.")
else:
    print("커피를 마시기에 돈이 부족하다.")
```

```python
# "5000원 이상의 현금을 가지고 있거나 카드가 있으면 택시를 타고 그렇지 않으면 걸어 가라."

cash = 4000
card = True

if cash >= 5000 or card:
    print("take a taxi")
else:
    print("take a walk")
```

## while 문

```python
# 열번 찍어 안넘어가는 나무 없다. 

tree_chop = 0

while tree_chop < 10:
    tree_chop = tree_chop + 1
    print(f"나무를 {tree_chop}번 찍었습니다.")
    if tree_chop == 10:
        print("나무가 넘어간다.")
```

```python
# 커피 머신 작동을 조건문으로 작성

money = 300
coffee = 10

while money:
    print("돈을 받았으니 커피를 줍니다.")
    coffee = coffee - 1
    print(f"남은 커피량은 {coffee}개 입니다.")
    if coffee == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지합니다.")
        break
```

```python
# if_else

iris = sns.load_dataset("iris")
iris["group"] = np.where(iris['species'] == 'setosa', "1", "2")
iris.iloc[[0, 1, 148, 149], :]
```

```python
iris["sepal"] = np.where(iris["sepal_length"] >= iris["sepal_length"].mean(), 1, 0)
iris.iloc[[0, 1, 148, 149], :]
```

```python
# if_elif_else : 방법 1

iris["sepal2"] = np.where(iris["sepal_length"] < 5, "small", np.where((iris["sepal_length"] >= 5) & (iris["sepal_length"] < 6), "medium", "large"))

iris["sepal2"].value_counts()

# def make_var(sepal_length):
#     if sepal_length < 5:
#         return "small"
#     elif (sepal_length >= 5) & (sepal_length < 6):
#         return "medium"
#     else:
#         return "large"
    
# iris["sepal3"] = iris["sepal_length"].map(make_var)

# iris.head()
 
```

```python
# if_elif_else : 방법 2

condition = [
    (iris["sepal_length"].lt(5)),
    (iris["sepal_length"].ge(5) & iris["sepal_length"].lt(6)),
    (iris["sepal_length"].ge(6)) 
     ]

choice = ["small", "medium", "large"]

iris["size"] = np.select(condition, choice)

iris["size"].value_counts()
```

```python
# if_elif_else : 방법 3

iris["sepal3"] = iris["sepal_length"].map(lambda x: "small" if x < 5 else ("medium" if 5 <= x < 6 else "large"))

iris["sepal3"].value_counts()
```

```python
# 연산을 선택하고 싶을 때

def add_mul(choice, *args):
    if choice == "add": 
        result = 0 
        for i in args: 
            result = result + i 
    elif choice == "mul": 
        result = 1 
        for i in args: 
            result = result * i 
    return result

add_mul("add", 1,2,3)
add_mul("mul", 1,3,5)
```

## 다른 예를 만들어 봅니다.

```python
# FourCal 클래스를 만들고 4개의 함수를 작성하여 계산하는 경우

class FourCal:
    def __init__(self, first, second):
        self.first = first
        self.second = second
        
    def setdata(self, first, second):
        self.first = first
        self.second = second
        
    def add(self):
        result = self.first + self.second
        return result
    
    def mul(self):
        result = self.first * self.second
        return result
    
    def sub(self):
        result = self.first - self.second
        return result
    
    def div(self):
        result = self.first / self.second
        return result
    
a = FourCal(3, 7)

a.add()
a.mul()
a.sub()
a.div()
```

```python
# 상속

class MoreFourCal(FourCal):
    def pow(self):
        result = self.first ** self.second
        return result

a = MoreFourCal(2, 3)
a.pow()
```
